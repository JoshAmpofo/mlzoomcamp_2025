```python
import onnxruntime as ort
```

```python
onnx_model_path = "clothing-model-new.onnx"
session = ort.InferenceSession(onnx_model_path, providers=["CPUExecutionProvider"])
```

```python
inputs = session.get_inputs()
input_name = inputs[0].name
print("Input name:", input_name)

outputs = session.get_outputs()
output_name = outputs[0].name
print("Output name:", output_name)
```

```python
# get image
url = "http://bit.ly/mlbookcamp-pants"
```

```python
!pip install keras-image-helper 
```

```python
# import keras_image_helper
from keras_image_helper import create_preprocessor
preprocessor = create_preprocessor('xception', target_size=(299,299))
```

```python
X = preprocessor.from_url(url)
```

```python
X
```

```python
preds = session.run([output_name], {input_name: X})
```

```python
float_preds = preds[0][0].tolist() # to get the predictions for a single image
```

```python
classes = [
    "dress",
    "hat",
    "longsleeve",
    "outwear",
    "pants",
    "shirt",
    "shoes",
    "shorts",
    "skirt",
    "t-shirt"
]

dict(zip(classes, float_preds))
```
